```python
from typing import ValuesView
import csv
with open('/content/9606_Q2_dup - 9606_Q2.csv') as f:
    observations=[tuple(line) for line in csv.reader(f)]

#print(len(observations))

#then creates an event link dictionary. i.e something that link every event to all its mnbrs
eventLinks = {}

for (Gene1, Gene2) in observations :
    #If this event have never been encoutered then create a new entry in links
    if not Gene1 in eventLinks.keys():
        eventLinks[Gene1] = []

    eventLinks[Gene1].append(Gene2)
#print(eventLinks)
#print(len(eventLinks.keys()))


#import pandas as pd
#df_keys= pd.DataFrame(eventLinks.keys())
#df_keys.to_csv('Keys.csv', index=False)

eventLinks.pop('GeneName_A')

print(len(eventLinks.keys()))


#print(eventLinks['TFPT'])

adjacencyList = eventLinks
#for key in adjacencyList.keys():
 #          for value in adjacencyList[key]:
              # print(value)

import pandas as pd

# creating a data frame
df = pd.read_csv('/content/9606_Q2_dup - 9606_Q2.csv')
print(df.head())

SetOfGene2=set(df['GeneName_B'])
#print(SetOfGene2)

#df_keys= pd.DataFrame(SetOfGene2)
#df_keys.to_csv('Columns.csv', index=False)

SetOfGene2 = {x for x in SetOfGene2 if x==x}
print(SetOfGene2)

#import pandas as pd
df = pd.DataFrame(0,index= adjacencyList.keys(),columns=SetOfGene2)


print(df.head())
#df.to_csv('/content/ZeroData.csv')
for rowIndex, row in df.iterrows(): #iterate over rows
   for i in eventLinks[rowIndex]:
        df.at[rowIndex,i]= 1

#for columnIndex, value in row.items():
 # df[columnIndex] = df[columnIndex].fillna(0)

#print(df) 
df.info()
df.to_csv('/content/AdjMatrix.csv')
```
